# Load Day-Ahead Auction Data (DAA)

```python
import pandas as pd
```

```python
daa = pd.read_csv('Gro_handelspreise_202401010000_202501010000_Viertelstunde.csv', 
                     sep=';', 
                     decimal=',',
                     usecols=[2],
                     # header=None,
                     # index_col=0
                )
```

```python
daa.rename(columns={'Deutschland/Luxemburg [€/MWh] Berechnete Auflösungen': 'DAA'}, inplace=True)
```

```python
daa.index = pd.date_range(start='2024-1-1 00:00', 
                         end='2024-12-31 23:45', 
                         freq='15min ')
```

```python
daa.tail(20)
```

```python
daa.describe()
```

```python
rows_to_drop = daa.index[-192:]  # Get the last 3 index values
daa.drop(rows_to_drop, inplace=True)
daa.tail()
```

```python
daa.describe()
```

```python
daa.plot()
```

```python
daa.to_csv('2024_daa.csv')
```
