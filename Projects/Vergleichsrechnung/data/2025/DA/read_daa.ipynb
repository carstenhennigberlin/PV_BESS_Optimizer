# Load Day-Ahead Auction Data (DAA)

```python
import pandas as pd
```

```python
daa = pd.read_csv('Gro_handelspreise_202412300000_202512290000_Viertelstunde.csv', 
                     sep=';', 
                     decimal=',',
                     usecols=[2],
                     # header=None,
                     # index_col=0
                )
```

```python
daa.rename(columns={'Deutschland/Luxemburg [€/MWh] Berechnete Auflösungen': 'DAA'}, inplace=True)
```

```python
daa.index = pd.date_range(start='2024-12-30 00:00', 
                         end='2025-12-28 23:45', 
                         freq='15min ')
```
