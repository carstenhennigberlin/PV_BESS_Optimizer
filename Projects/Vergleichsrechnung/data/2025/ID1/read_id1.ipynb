# Load Intraday Continous Data (ID1)

```python
import pandas as pd
```

```python
dfs = []
for i in range(1, 53):
    df = pd.read_csv(f'energy-charts_Stromproduktion_und_Börsenstrompreise_in_Deutschland_in_Woche_{i}_2025.csv', 
                     sep=';', 
                     decimal=',',
                     header=None,
                     index_col=0)
    
    df.sort_index(inplace=True)                
    dfs.append(df)
```

```python
id1 = pd.concat(dfs)
```

```python
id1.head()
```

```python
id1.rename(columns={1: 'ID1'}, inplace=True)
id1.index = pd.date_range(start='2024-12-30 00:00', 
                         end='2025-12-28 23:45', 
                         freq='15min ')
```

```python
id1.describe()
id1.head()
```

```python
id1.plot()
```

```python
id1.to_csv('2025_id1.csv')
```
